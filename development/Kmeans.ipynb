```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
```

```python
df = pd.read_csv(os.path.join(".", "datasets", "winequality-red.csv"))
df.head()
```

```python
y = df["quality"]
y
```

```python
X = df.drop("quality", axis=1)
X.head()
```

```python
print(f"Labels: {y[:10]}")
print(f"Data: {X[:10]}")
```

```python
# Visualizing both classes
#plt.scatter(X[:, 0], X[:, 1], c=y)
y_arr = y.to_numpy()
y_arr
```

```python
X_arr = X.to_numpy()
X_arr
```

```python
X=X_arr
X
```

```python
y=y_arr
y
```

```python
# Plot and show scatter
#plt.scatter(X[:, 0], X[:, 1])
import numpy as np
plt.scatter(X[:, 1], X[:, 8], X[: 10], y)
plt.show()
```

```python
# Create a kmeans model using k = 12
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=7)

# Fit the model to the data
kmeans.fit(X)

# Use the data to predict the clusters
# save the predictions as `predicted_clusters`
predicted_clusters = kmeans.predict(X)
```

```python
# Plot the clusters
#plt.scatter(X[:, 0], X[:, 1], c=predicted_clusters, s=50, cmap='viridis')
plt.scatter(X[:, 1], X[:, 8], c=predicted_clusters, s=50, cmap='viridis')
plt.show()
```

```python
# Print the cluster centers and cluster labels
centers = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
centers
```

```python
plt.scatter(X[:, 1], X[:, 8], c=labels, s=20, cmap='viridis')
plt.scatter(centers[:, 1], centers[:, 8], c='red', s=200, alpha=0.5)

plt.show()
```
